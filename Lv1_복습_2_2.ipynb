## Í≤∞Ï∏°Ïπò Ï†ÑÏ≤òÎ¶¨
---

```python
# dropna() Î•º ÏÇ¨Ïö©Ìï¥ train Îç∞Ïù¥ÌÑ∞Îäî Í≤∞Ï∏°ÏπòÎ•º Ï†úÍ±∞ÌïòÍ≥†      Í≤∞Ï∏°Ïù¥ ÏûàÎäî ÌñâÏù¥ ÏÇ¨ÎùºÏßÑÎã§.
# fillna() Î•º ÏÇ¨Ïö©Ìï¥ test Îç∞Ïù¥ÌÑ∞ Í≤∞Ï∏°ÏπòÎäî 0 ÏúºÎ°ú ÎåÄÏ≤¥ÌïòÏÑ∏Ïöî.
# Í∑∏Î¶¨Í≥† Í≤∞Ï∏°ÏπòÏùò Í∞ØÏàòÎ•º Ï∂úÎ†•ÌïòÏó¨ ÌôïÏïàÌïòÏÑ∏Ïöî.                    # Ïã§Ïàò train.dropna() ÎÇò fillna(Î≠êÏãúÍ∏∞) ÌïúÎã§Í≥† ÎåÄÏ≤¥ÎêòÎäî Í≤ÉÏù¥ ÏïÑÎãò. ÎåÄÏûÖÏùÑ Ìï¥Ï£ºÏñ¥Ïïº ÌïúÎã§.
train = train.dropna()
test = test.fillna(0)

print(train.isnull().sum())
```

```python
train.shape
```

## Î™®Îç∏ ÌõàÎ†®
---


```python
# train Îç∞Ïù¥ÌÑ∞Ïùò count ÌîºÏ≥êÎ•º Î∫Ä Í≤ÉÏùÑ X_train ÏúºÎ°ú Ìï†ÎãπÌïòÏÑ∏Ïöî.
# train Îç∞Ïù¥ÌÑ∞Ïùò count ÌîºÏ≥êÎßåÏùÑ Í∞ÄÏßÑ Í≤ÉÏùÑ Y_train ÏúºÎ°ú Ìï†ÎãπÌïòÏÑ∏Ïöî.
# ÌöåÍ∑ÄÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º ÏÑ†Ïñ∏ÌïòÍ≥† fit() ÏúºÎ°ú ÌõàÎ†®ÏãúÌÇ§ÏÑ∏Ïöî.
```

```python
X_train = train.drop(['count'],axis=1)   # axis =0 Ïóê ÎåÄÌïúÍ≤ÉÏùÄ 2Ï∞®Ïõê Î∞©Ìñ•Ïù¥ÎØÄÎ°ú Ïó¥Î∞©Ìñ• Ïù∏Îç±Ïä§Í∞íÏùÑ dropÌï† Îïå ÏÇ¨Ïö©, axis =1 Ïóê ÎåÄÌïúÍ≤ÉÏùÄ 1Ï∞®Ïõê Î∞©Ìñ•Ïù¥ÎØÄÎ°ú labelÏùÑ Ïù¥Ïö©
Y_train = train['count']  # Î∞òÎåÄÎ°ú indexÎ°ú ÎΩëÏïÑÎÇ¥Í∏∞ => train[train.index==ÎÇ¥Í∞Ä ÏõêÌïòÎäî Ïù∏Îç±Ïä§ Í∞í]
```

## ÌÖåÏä§Ìä∏ ÌååÏùº ÏòàÏ∏°
---

```python
# predict() ÏùÑ Ïù¥Ïö©Ìï¥ test data Î•º ÌõàÎ†®Ìïú Î™®Îç∏Î°ú ÏòàÏ∏°Ìïú data array Î•º ÏÉùÏÑ±ÌïòÏÑ∏Ïöî.
model = DecisionTreeRegressor()  #ÌöåÍ∑Ä
model.fit(X_train,Y_train)
model2 = DecisionTreeClassifier()  #Î∂ÑÎ•ò
model2.fit(X_train,Y_train)

pred = model.predict(test)
pred2 = model2.predict(test)
pred.shape
pred2.shape
```

## submission ÌååÏùº ÏÉùÏÑ±

```python
# submission.csv Î•º Î∂àÎü¨Ïò§ÏÑ∏Ïöî.
# submission df Ïùò count ÌîºÏ≥êÏóê ÏòàÏ∏°Ìïú Í≤∞Í≥ºÎ•º ÎçßÏûÖÌûàÏÑ∏Ïöî.
# submission df Î•º to_csv() Î•º Ïù¥Ïö©Ìï¥ csv ÏùÑ ÏÉùÏÑ±ÌïòÏÑ∏Ïöî. *index=False) 
submission = pd.read_csv('data/submission.csv')
submission['count'] = pred
submission.to_csv('sub.csv',index=False)
```
